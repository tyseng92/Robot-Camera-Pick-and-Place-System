```python
def map_id(idx):

    with open(target_path) as t:
        with open(coco_path) as c:
            tlist = t.read().splitlines()
            clist = c.read().splitlines()
            print(tlist)
            print(clist)
            mapping = {index:tlist.index(value) if value in tlist else 'NA' for index, value in enumerate(clist)}
            mapping = collections.OrderedDict(sorted(mapping.items()))
            print(mapping)
    return str(mapping[int(idx)])
#num = int(map_id('74'))
```

```python
def copy_files(src, dst):
    for root, dirs, files in os.walk(src):  
       for file in files:
          path_file = os.path.join(root,file)
          shutil.copy2(path_file,dst) 
copy_files(src_img_path, dst_path)
copy_files(src_txt_path, dst_path)
```

```python
def main():
    copy_files()
    tpath = os.path.join(original_path, "*.txt")
    for fname in glob.glob(tpath):
        bname = os.path.basename(fname)
        dname = os.path.join(mapped_path, bname) 
        with open(fname) as f:
            flist = f.read().splitlines()
            delete(dname)
            for fl in flist:
                ulist = fl.split()
                ulist[0] = map_id(ulist[0])
                with open(dname, 'a') as f:
                    if ulist[0] != 'NA':
                        f.write(' '.join(ulist) + str('\n'))       
    print("Done")
```
